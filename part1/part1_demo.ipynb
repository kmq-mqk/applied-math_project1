**Cell 2: Test Gaussian Elimination**

```python
# Case 1: Normal
A = Matrix([
    [0, 2, 1],
    [1, -2, -3],
    [-1, 1, 2]
])
b = [-8, 0, 3]

result = A.gaussian_eliminate(b)
A_bar, x, count = result
print("NGHIỆM CỦA HỆ PHƯƠNG TRÌNH:", x)
verify_solution(A.data, x, b)


# Case 2: Pivot = 0 (Vô nghiệm)
A = Matrix([
    [1, 0, 0],
    [0, 0, 3],
    [0, 0, 1],
])
b = [3, 15, 14]

result = A.gaussian_eliminate(b)
A_bar, x, count = result
print("NGHIỆM CỦA HỆ PHƯƠNG TRÌNH:", x)
verify_solution(A.data, x, b)
```

**Cell 3: Test Determinant**

```python
A = Matrix([
    [0, 2, 1],
    [1, -2, -3],
    [-1, 1, 2]
])
custom_det = A.det()

print("--- KẾT QUẢ CUSTOM ---")
print(f'Định thức của ma trận A: {custom_det}')

print("\n--- KIỂM CHỨNG VỚI NUMPY ---")
np_det = np.linalg.det(np.array(A.data))
print(f'Định thức của ma trận A (NumPy): {np_det:.4f}')
```

**Cell 4: Test Inverse**

```python
m2 = Matrix([
    [1, 2, 3],
    [0, 1, 4],
    [5, 6, 0]
])
r2 = m2.inverse()

print("--- KẾT QUẢ CUSTOM ---")
print("Ma trận nghịch đảo A^-1:")
r2_list = r2.data if hasattr(r2, 'data') else r2
for row in r2_list:
    print(row)

print("\n--- KIỂM THỬ A * A^-1 = I ---")
np_m2 = np.array(m2.data)
np_r2 = np.array(r2_list)
I_check = np.dot(np_m2, np_r2)
print("Kết quả A * A^-1 (Làm tròn):")
print(np.round(I_check, 0))

print("\n--- KIỂM CHỨNG VỚI NUMPY ---")
np_inv = np.linalg.inv(np_m2)
print("Ma trận nghịch đảo A^-1 (NumPy):")
print(np_inv)
```
