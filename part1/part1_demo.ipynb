**Cell 2: Test Gaussian Elimination**

```python
# --- DỮ LIỆU KIỂM THỬ (TEST CASES) ---
test_data = [
    {
        "A": [[0, 2, 1], [1, -2, -3], [-1, 1, 2]],
        "b": [-8, 0, 3],
        "desc": "Case 1: Normal (Swap row)"
    },
    {
        "A": [[1, 0, 0], [0, 0, 3], [0, 0, 1]],
        "b": [3, 15, 14],
        "desc": "Case 2: Pivot = 0 (No Solution)"
    },
    {
        "A": [[1, 1, 1], [0, 2, -1], [1, 3, 0]],
        "b": [6, 3, 9],
        "desc": "Case 3: Infinitely Many Solutions"
    },
    {
        "A": [[1, 0, 0], [0, 1, 0], [0, 0, 1]],
        "b": [5, -2, 10],
        "desc": "Case 4: Identity Matrix"
    },
    {
        "A": [[2, 4, -2], [0, 1, 1], [0, 0, 4]],
        "b": [2, 4, 8],
        "desc": "Case 5: Upper Triangular"
    }
]

# --- CHẠY KIỂM THỬ ---
for i, test in enumerate(test_data):
    print(f"--- {test['desc']} ---") 
    
    A_obj = Matrix(test['A'])
    b_val = test['b']
    
    # Thực hiện khử Gauss
    result = A_obj.gaussian_eliminate(b_val)
    A_bar, x, count = result
    
    # In kết quả nghiệm
    print(f"NGHIỆM CỦA HỆ PHƯƠNG TRÌNH: {x}")
    
    if x is not None:
        verify_solution(test['A'], x, b_val)
    else:
        print("Solution is true")
    
    print() # Tạo dòng trống giữa các case cho dễ nhìn
```

**Cell 3: Test Determinant**

```python
test_det_data_part1 = [
    {"A": [[0, 2, 1], [1, -2, -3], [-1, 1, 2]]}, # Case 1: Swap row
    {"A": [[1, 2, 3], [2, 4, 6], [7, 8, 9]]},    # Case 2: Det = 0
    {"A": [[1, 0, 0], [0, 1, 0], [0, 0, 1]]},    # Case 3: Identity
    {"A": [[2, 0, 0], [0, 5, 0], [0, 4, 3]]}     # Case 4: Triangular
]

for test in test_det_data_part1:
    A_obj = Matrix(test["A"])
    custom_det = float(A_obj.det()) + 0.0
    
    print("--- KẾT QUẢ CUSTOM ---")
    print(f'Định thức của ma trận A: {custom_det:.4f}')

    print("\n--- KIỂM CHỨNG VỚI NUMPY ---")
    np_det = np.linalg.det(np.array(A_obj.data)) + 0.0
    print(f'Định thức của ma trận A (NumPy): {np_det:.4f}\n')
```

**Cell 4: Test Inverse**

```python
test_inv_1_2 = [
    {"A": [[1, 2, 3], [0, 1, 4], [5, 6, 0]]},
    {"A": [[1, 0, 0], [0, 1, 0], [0, 0, 1]]}
]

for test in test_inv_1_2:
    m = Matrix(test["A"])
    r = m.inverse()
    r_list = r.data if hasattr(r, 'data') else r
    
    print("--- KẾT QUẢ CUSTOM ---")
    for row in r_list:
        print([round(val, 4) for val in row])

    print("\n--- KIỂM THỬ A * A^-1 = I ---")
    print(np.round(np.dot(np.array(m.data), np.array(r_list)), 0) + 0.0)
    print("-" * 30)
```
